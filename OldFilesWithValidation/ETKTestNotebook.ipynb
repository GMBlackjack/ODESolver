So now the great plan is to *create a thorn*. This thorn is going to start out rather simple with an easy goal: just run the differential equation solver and output the results to a file. Let's see how this goes... we decided to do the files manually rather than with a notebook, as drag'n'drop is superior to command line interface. 

We just need to manually add it to the thorn list. 

```python
import os
thorn_list="/home/gsteward/Desktop/Cactus/configs/sim/ThornList"
new_thorn_path = "EinsteinInitialData/TOVOdieGM"
def add_thorn(new_thorn_path):
    contents = None
    with open(thorn_list, 'r') as f:
        contents = f.read()
    if new_thorn_path not in contents:
        with open(thorn_list, 'a') as f:
            f.write(new_thorn_path + '\n')
add_thorn(new_thorn_path)
!tail {thorn_list}
```
